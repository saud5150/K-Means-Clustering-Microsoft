### Importing Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
from sklearn.cluster import KMeans
```

Loading the data

```python
data = pd.read_csv('Iris.csv');

data
```

Plotting the data

```python
plt.scatter(data['sepal_length'], data['petal_length'])
plt.show()
```

```python
x = data.iloc[:, :]
x
```

Clustering

```python
km = KMeans(4)

km.fit(x)
```

Clustering results


```python
df = km.fit_predict(x)
```

```python
newdf = data.copy()
newdf['pred'] = df
newdf
```

Plot the data

```python
plt.scatter(newdf['sepal_length'], newdf['petal_length'], c = newdf['pred'], cmap = 'rainbow')
plt.show()
```
